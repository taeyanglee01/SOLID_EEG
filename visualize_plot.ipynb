```python
step = [1000, 10000, 20000, 30000, 40000, 50000,
        60000, 70000, 80000, 90000, 100000]

ratio09_MSE = [40.831, 10.413, 9.594, 9.527, 9.002, 8.906, 
               8.153, 8.376, 8.313, 8.112, 8.104]

ratio07_MSE = [41.48, 17.249, 16.702, 16.818, 14.592, 15.010,
               14.302, 14.140, 14.494, 14.159, 14.344]

ratio05_MSE = [42.251, 23.688, 22.317, 22.147, 20.715, 21.317, 
               19.840, 20.228, 20.718, 20.912, 20.722]
```

```python
# Keep ratio comparison with train steps

plt.figure(figsize=(6, 4), dpi=300)

plt.plot(step, ratio09_MSE, marker='o', linewidth=2,
         color='#1f77b4', label='Keep Ratio 0.9')
plt.plot(step, ratio07_MSE, marker='s', linewidth=2,
         color='#2ca02c', label='Keep Ratio 0.7')
plt.plot(step, ratio05_MSE, marker='^', linewidth=2,
         color='#d62728', label='Keep Ratio 0.5')

# plt.xscale('log')   # 
plt.xlabel('Training Steps')
plt.ylabel('MSE')
plt.legend(frameon=False)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

keep_ratio = ['100%', '90%', '70%', '50%']
clip_true = [1.116, 6.152, 11.541, 16.955]
clip_false = [51.180, 60.210, 67.901, 84.282]

x = np.arange(len(keep_ratio))
width = 0.35

plt.figure(figsize=(6, 4), dpi=300)

plt.bar(x - width/2, clip_true, width,
        label='Clip True', color='#6baed6')   #  
plt.bar(x + width/2, clip_false, width,
        label='Clip False', color='#fc9272')  #  

plt.xlabel('Keep Ratio')
plt.ylabel('MSE')
plt.xticks(x, keep_ratio)
plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data (90% keep ratio comparison)
methods = ['tanh scaling', 'z divide 3']
mse_values = [6.152, 15.735]

x = np.arange(len(methods))

plt.figure(figsize=(4, 4), dpi=300)

bars = plt.bar(
    x, mse_values, width=0.5,              # thinner bars
    color=['#d9f0a3', '#a1d99b']             #  ,  
)

plt.ylabel('MSE')
plt.xticks(x, methods)

# value annotation on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:.3f}',
        ha='center',
        va='bottom',
        fontsize=8
    )

plt.tight_layout()
plt.show()
```
